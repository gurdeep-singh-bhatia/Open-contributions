# Image Blending 

We know that images are saved in computer as matrix of pixels. So if we add two images it means we are adding two matrices.
And we can do this by opencv add() function 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
img1=cv2.imread("Desktop/lena.jpg")
```

```python
print(img1)
```

```python
img1.shape
```

```python
plt.imshow(img1)
```

```python
cv2.imshow("img1",img1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img2=cv2.imread("Desktop/rocket.jpg")
```

```python
img2.shape
```

```python
plt.imshow(img2)
```

```python
cv2.imshow("img2",img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

Now for adding two images we have to resize both the images so that they have equal dimensions 

```python
img1=cv2.resize(img1,(500,500),interpolation=cv2.INTER_AREA)
```

```python
img2=cv2.resize(img2,(500,500),interpolation=cv2.INTER_AREA)
```

```python
added_img=cv2.add(img1,img2)
```

```python
plt.imshow(added_img)
```

As we can see that two images get added 

```python
cv2.imshow("added_img",added_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Now sometimes we dont want only adding two images, sometimes we want that different image will get different weithage or diff weights so that we get some transparency so this is called blending and we can do this by opencv addWeighted function 
This is the equation of blending (special adding ):

g(x) = (1 - a)f(x) + af1(x)

```python
blend_img = cv2.addWeighted(img1, 0.3, img2, 0.7, 0) 
```

```python
plt.imshow(blend_img)
```

```python
blend_img = cv2.addWeighted(img2, 0.3, img1, 0.7, 0) 
```

```python
plt.imshow(blend_img)
```

### We can also blend our image with its blurred version lets us see

```python
processed_image=cv2.addWeighted (img1,4, cv2.GaussianBlur( img1, (0,0) , 250/10) ,-4 ,128) # using gaussian blurring , u can use any 
```

```python
plt.imshow(processed_image)
```

```python
processed_image=cv2.addWeighted (img2,4, cv2.GaussianBlur( img2, (0,0) , 250/10) ,-4 ,128) # using gaussian blurring , u can use any 
```

```python
plt.imshow(processed_image)
```
